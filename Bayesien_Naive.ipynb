```python
import math
import numpy as np
import pandas as pd
from csv import reader
from statistics import *
```

```python
#--------------------------------------------------------
data_sexe = pd.read_csv("sexEchantillon.csv", sep=";")
#--------------------------------------------------------
# Détermination Pféminin et Pmasculin
Pop = data_sexe.groupby('Sexe')['Effectif'].count()

effectif_F = Pop[0]
effectif_M = Pop[1]
totalEffectif = effectif_M + effectif_F

#Prob_M = effectif_M / totalEffectif
#Prob_F = effectif_F / totalEffectif
#--------------------------------------------------------
# Filtre des données par genres
df_sexe_masculin = data_sexe['Sexe'] == 'Masculin'
filtre_sexe_M = data_sexe[df_sexe_masculin]

df_sexe_feminin = data_sexe['Sexe'] == 'Feminin'
filtre_sexe_F = data_sexe[df_sexe_feminin]
#--------------------------------------------------------
# Création des listes de valeurs de données par genre
Taille_M = filtre_sexe_M['Taille']
Poids_M = filtre_sexe_M['Poids']
Pointure_M = filtre_sexe_M['Pointure']

Taille_F = filtre_sexe_F['Taille']
Poids_F = filtre_sexe_F['Poids']
Pointure_F = filtre_sexe_F['Pointure']
#--------------------------------------------------------
#Tableau des Espérances et variances
Esp_Taille_M = mean(Taille_M)
Var_Taille_M = variance(Taille_M)
Esp_Poids_M = mean(Poids_M)
Var_Poids_M = variance(Poids_M)
Esp_Pointure_M = mean(Pointure_M)
Var_Pointure_M = variance(Pointure_M)

Esp_Taille_F = mean(Taille_F)
Var_Taille_F = variance(Taille_F)
Esp_Poids_F = mean(Poids_F)
Var_Poids_F = variance(Poids_F)
Esp_Pointure_F = mean(Pointure_F)
Var_Pointure_F = variance(Pointure_F)

TB_Esp_Var = pd.DataFrame({'EsperanceTaille':[Esp_Taille_M,Esp_Taille_F],
                           'VarianceTaille':[Var_Taille_M,Var_Taille_F],
                           'EsperancePoids':[Esp_Poids_M,Esp_Poids_F],
                           'VariancePoids':[Var_Poids_M,Var_Poids_F],
                           'EsperancePointure':[Esp_Pointure_M,Esp_Pointure_F],
                           'VariancePointure':[Var_Pointure_M,Var_Pointure_F]},
                          index=['Masculin','Feminin'])
#--------------------------------------------------------
#Echantillon à classifier, sexe inconnu
Unknown_Taille = 183
Unknown_Poids = 59
Unknown_Pointure = 20

Unknown = pd.DataFrame({'Taille':[Unknown_Taille],'Poids':[Unknown_Poids],'Pointure':[Unknown_Pointure]})
#--------------------------------------------------------
# Calcul des probabilités
P_masculin = effectif_M/totalEffectif
P_taille_masculin = 1/(math.sqrt((2*math.pi)*Var_Taille_M))*math.exp(-((Unknown_Taille-Esp_Taille_M)**2)/(2*Var_Taille_M))
P_poids_masculin = 1/(math.sqrt((2*math.pi)*Var_Poids_M))*math.exp(-((Unknown_Poids-Esp_Poids_M)**2)/(2*Var_Poids_M))
P_pointure_masculin = 1/(math.sqrt((2*math.pi)*Var_Pointure_M))*math.exp(-((Unknown_Pointure-Esp_Pointure_M)** 2)/(2*Var_Pointure_M))
Pp_masculin = P_masculin*P_taille_masculin*P_poids_masculin*P_pointure_masculin

P_feminin = effectif_F/totalEffectif
P_taille_feminin = 1/(math.sqrt((2*math.pi)*Var_Taille_F))*math.exp(-((Unknown_Taille-Esp_Taille_F)**2)/(2*Var_Taille_F))
P_poids_feminin = 1/(math.sqrt((2*math.pi)*Var_Poids_F))*math.exp(-((Unknown_Poids-Esp_Poids_F)**2)/(2*Var_Poids_F))
P_pointure_feminin = 1/(math.sqrt((2*math.pi)*Var_Pointure_F))*math.exp(-((Unknown_Pointure-Esp_Pointure_F)** 2)/(2*Var_Pointure_F))
Pp_feminin = P_feminin*P_taille_feminin*P_poids_feminin*P_pointure_feminin

TB_Proba_Unknown = pd.DataFrame({'Homme':[P_masculin,P_taille_masculin,P_poids_masculin,P_pointure_masculin,Pp_masculin],
                                 'Femme':[P_feminin,P_taille_feminin,P_poids_feminin,P_pointure_feminin,Pp_feminin]},
                               index = ['P_pop','P_taille','P_poids','P_pointure','Pp'])

TB_Esp_Var
Unknown
TB_Proba_Unknown
```
